```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
```

```python
fp='Sacramentorealestatetransactions.csv'
df=pd.read_csv(fp,header=0)
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['baths'],df['price'])
```

```python
plt.scatter(df['beds'],df['price'])
```

```python
plt.scatter(df['sq__ft'],df['price'])
```

```python
xpts=np.array(df['price'])
ypts=np.array(df['sq__ft'])
```

```python
reg=LinearRegression()
reg.fit(xpts.reshape(-1,1), ypts.reshape(-1,1))
print(reg.coef_)
print(reg.intercept_)
```

```python
fig=plt.scatter(xpts,ypts)
y=reg.coef_*xpts+reg.intercept_
fig=plt.plot(xpts,y.T,lw=4,c='orange',label='Regression line')
plt.xlabel('price')
plt.ylabel('sq__ft')
```

```python
yhat=reg.predict(xpts.reshape(-1,1))
fig=plt.scatter(ypts,yhat)
ab=ordi=range(0,5000)
fig=plt.plot(ab,ordi,lw=4,c='orange',label='45.degree line')
plt.xlabel('Observations',size=14)
plt.xlabel('predictions',size=14)
```

```python
np.corrcoef(ypts.T,yhat.T)
```
