```python
x=np.arange(-2*math.pi,2*math.pi,0.01)
amps=np.sin(x)
ampc=np.cos(x)
```

```python
plt.plot(x,amps)
plt.plot(x,ampc)
plt.title('Sine and Cosine Waves')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(y=0,color='k')
plt.legend(['Sine','Cosine'],loc="lower left")
```

```python
fp='phbirths.csv'
df=pd.read_csv(fp,header=0)
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['gestate'],df['grams'])
```

```python
xpts=np.array(df['gestate'])
ypts=np.array(df['grams'])
```

```python
reg=LinearRegression()
reg.fit(xpts.reshape(-1,1), ypts.reshape(-1,1))
print(reg.coef_)
print(reg.intercept_)
```

```python
fig=plt.scatter(xpts,ypts)
y=reg.coef_*xpts+reg.intercept_
fig=plt.plot(xpts,y.T,lw=4,c='orange',label='Regression line')
plt.xlabel('Gestate')
plt.ylabel('Grams')
```

```python
yhat=reg.predict(xpts.reshape(-1,1))
fig=plt.scatter(ypts,yhat)
ab=ordi=range(0,5000)
fig=plt.plot(ab,ordi,lw=4,c='orange',label='45.degree line')
plt.xlabel('Observations',size=14)
plt.xlabel('predictions',size=14)
```

```python
np.corrcoef(ypts.T,yhat.T)
```
